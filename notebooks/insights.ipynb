## Homicidios por año

```python
df["Año del hecho"].value_counts().sort_index().plot(kind="line", marker="o")
```

Cambio importante:
- A partir de 2021 ocurre un aumento fuerte y sostenido.
- Se pasa de aproximadamente 11,300 en 2020 a más de 13,000 en 2021.

Pico maximo:
- El punto más alto es 2023 (~14,200 casos)
- Luego en 2024 hay una leve caída, pero sigue alto

# Conclusiones

El análisis de los homicidios permite identificar patrones claros en la ocurrencia de la violencia:


- La mayoría de los homicidios ocurre en zonas urbanas (cabeceras municipales), lo que indica una alta concentración de la violencia en contextos urbanos.
- El escenario principal es la vía pública, lo que sugiere que la violencia es predominantemente visible y ocurre en espacios abiertos.
- Existe una fuerte diferencia por sexo: los hombres representan la gran mayoría de las víctimas, lo que evidencia una mayor exposición a dinámicas de violencia asociadas a entornos urbanos y sociales.
- Los homicidios no ocurren únicamente en contextos delictivos, sino también durante actividades cotidianas como el tiempo libre o la asistencia a eventos, lo que indica que la violencia puede surgir en situaciones comunes.
- A lo largo del tiempo, se observa una tendencia creciente en los últimos años, especialmente después de 2020.

En conjunto, estos resultados sugieren que la violencia homicida está asociada principalmente a dinámicas urbanas, espacios públicos y población masculina.

# Limitaciones del dataset

Durante el análisis se identificaron varias limitaciones que afectan la calidad e interpretación de los datos:

- Existe una alta proporción de registros en categorías como “Otra” y “Sin información”, especialmente en variables como “Actividad Durante el Hecho”, lo que reduce la precisión del análisis.
- Algunas variables presentan altos porcentajes de valores nulos, por ejemplo:
Orientación Sexual (~66%)
Identidad de Género (~66%)
Transgénero (~66%)
- Se encontraron inconsistencias en los datos categóricos, como:
Diferencias en mayúsculas/minúsculas
Espacios adicionales
Variaciones en la escritura de una misma categoría
(ej: “Vía pública” vs “vía Pública”)
- Existen categorías duplicadas o mal definidas, especialmente en variables como:
Escenario del Hecho
Actividad Durante el Hecho

Estas limitaciones indican que, aunque el dataset permite identificar tendencias generales, los resultados deben interpretarse con cautela.
