Informacion que sacamos del EDA:

Tenemos 4 problemas principales:

1. Inconsistencias en texto:
- Soltero(a) vs Soltero (a)
- Unión libre vs Unión Libre
- Vía Pública vs Vía pública
- categorías repetidas con puntos o mayúsculas
Esto hace que se rompa el analisis


2. Sin informacion:
- Estado Civil → 20k
- Actividad → 52k 
- Zona → 1.5k
No son exactamente nulls pero por funcionalidad se tomaran como nulls.

3. Columnas con muchos nulls
- Orientación Sexual
- Identidad de Género
- Transgénero
Pueden sesgar el analisis.

4. Columnas inutules o constantes:
- Manera de muerte = 100% "Presuntos Homicidios"
No aporta nada.
